```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('Panama.csv')
```

```python
y = data['nat_demand']
date = data['datetime']
```

```python
plt.scatter(date,y)
plt.x_label('Date to time', fontsize = 10)
plt.y_label('Electricity demand', fontsize = 10)
```

```python
# 2. Convert the first column to datetime objects
data['Date_Time'] = pd.to_datetime(data.iloc[:, 0]) 

# 3. Set the Date_Time column as the index
data.set_index('Date_Time', inplace=True)

# 4. Choose the columns you want to visualize
columns_to_plot = data.columns[1:] # Exclude the first column (Date_Time)

# 5. Create subplots for each column
fig, axes = plt.subplots(len(columns_to_plot), 1, figsize=(12, 6 * len(columns_to_plot)))

# 6. Plot each column against Date_Time
for i, column in enumerate(columns_to_plot):
    axes[i].plot(data.index, data[column])
    axes[i].set_ylabel(column)
    axes[i].set_xlabel('Date and Time')

# 7. Adjust spacing between subplots
plt.tight_layout()

# 8. Display the plots
plt.show()
```
